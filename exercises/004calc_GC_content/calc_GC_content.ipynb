```python
with open("rosalind_gc.txt", "r") as rf:
    list_of_lines = rf.readlines()
    for l in list_of_lines:
        list_of_lines[list_of_lines.index(l)] = l.replace("\n", "")
    print(list_of_lines)
    dna_seqs = {}
    for line in list_of_lines:
        if line[0] == ">":
            dna_seqs[line] = ""
            for x in list_of_lines[list_of_lines.index(line) + 1::]:
                if x[0] == ">":
                    break
                dna_seqs[line] += x
            "".join(dna_seqs[line])
            
print(dna_seqs)

res = 0
res_name = ""
for n, dna in dna_seqs.items():
    print(n + "\n" + dna)
    GC_cont = (dna.count("G") + dna.count("C"))/len(dna) * 100
    if GC_cont > res:
        res = GC_cont
        res_name = n[1::]

with open("GC_cont.txt", "w") as wf:
    wf.write(res_name + "\n" + str(res))
```
